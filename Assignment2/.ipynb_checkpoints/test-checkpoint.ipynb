```python
class kNN:
    def __init__(self):                                        # initialize attributes to None
        self.imputation = None
        self.normalization = None
        
        self.labels = None
        self.training_labels = None
        self.training_data = None

    def fit(self, input_df, normalizationtype = "zscore"):    #Assign values
        df = input_df.copy()
        df, self.imputation = create_imputation(df)
        df, self.normalization = create_normalization(df, normalizationtype = "zscore")
        
        # self.training_labels should be a pandas series corresponding to the "CLASS" column, set to be of type "category" 
        self.training_labels = df.loc[:,"CLASS"].astype("category")
        
        # self.labels should be the categories of the previous series
        self.labels = [i for i in range(1, np.max(df.loc[:, "CLASS"])+1)]
        
        # self.training_data should be the values (an ndarray) of the transformed dataframe, i.e., after employing imputation, 
        df = df.drop(labels = ["CLASS", "ID"], axis = 1)
        self.training_data = df
        
    def get_KNN_prediction(self, vl, k):
        df = self.training_data                                      # import training data
        mat = df.values                                              # get values in training data
        data_size = df.shape[0]                                      # calculate the data size
        matVect = np.tile(vl, data_size).reshape(data_size,len(vl))  # assemble a matrix to calculate diff later
        sqDiff = (mat-matVect)**2                                   # calulate the difference 
        sqDistances = sqDiff.sum(axis = 1)
        distances = sqDistances**0.5                                # calculate the Euclid distance
        KNNIndex = distances.argsort()[:k]                          # return the index of nearest K neighbors
        k = KNNIndex.shape[0] 
        labels_data = self.training_labels                          #import neghbors
        labels = self.labels
        prob = np.zeros(len(labels))                                # initialize
        for i in range(k):                                         # counting class prob
            label_vect = np.array((labels == labels_data[KNNIndex[i]]), dtype='int64') 
            prob += label_vect
        prob = prob / k
        return(prob)
    
    def predict(self, input_df, k):
        df = input_df.copy()
        data_size = df.shape[0]
        probs = pd.DataFrame(columns = self.labels)
        df = df.drop(labels = ["CLASS", "ID"], axis = 1)           # drop "CLASS"&"ID" before applying functions
        df = apply_imputation(df, self.imputation)                 # data preprocessing
        df = apply_normalization(df, self.normalization)
        
        # Using KNN function defined to make prediction
        Mat = np.array([np.zeros(len(self.labels)) for i in range (data_size)])
        for i in range(data_size):
            df1 = df.iloc[i,:]
            vl = df1.values
            prob = self.get_KNN_prediction(vl = vl ,k = k)
            Mat[i] = prob
        probs = pd.DataFrame(Mat, columns=self.labels)
        return(probs)
```
